```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin1')
df.head()
```

```python
# create an array based on multiple price tiers (+2 choices)

# create 'conditions' and 'values' variables

conditions = [
    df['Price_euros'] > 3000,
    (df['Price_euros'] > 2000) & (df['Price_euros'] <= 3000),
    (df['Price_euros'] > 800) & (df['Price_euros'] <= 2000),
    df['Price_euros'] <= 800
]

values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set it to a new column

df['Price_tiers'] = np.select(conditions, values, default='Unknown')
```

```python
# count values inside the 'Price_tiers' column

df.value_counts('Price_tiers')
```

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big > 16, Big > 14, Small > 12, Too Small < 12

# create 'conditions' and 'values' variables

conditions2 = [
    (df['Inches'] > 16),
    (df['Inches'] > 14) & (df['Inches'] <= 16),
    (df['Inches'] > 12) & (df['Inches'] <= 14),
    df['Inches'] <= 12

]


values2 = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
# set it to a new column

df['Screen_size'] = np.select(conditions2, values2, default='Unknown')
```

```python
# count values inside 'Screen_size' column

df.value_counts('Screen_size')
```
