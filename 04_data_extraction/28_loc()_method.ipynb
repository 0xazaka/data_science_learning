### loc() 

is label-based, so we have to specify rows and columns based on their index/row and column labels

### iloc()

is integer position-based, so we have to specify rows and columns by their integer position values (0-based integer position)

```python
import pandas as pd
```

```python
# read csv file and show dataframe

df = pd.read_csv('players_20.csv')

df.head()
```

```python
# set index as player name and show dataframe

df = df.set_index('short_name')
df.head()
```

```python
# select columns

df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club', 'overall']]
df.head()
```

## Selecting elements by index label with .loc()
### 1. selecting with a single value
loc[row_label, column_label]

```python
# get all data about L.Messi

df.loc['L. Messi']
```

```python
# get the height of L.Messi

df.loc['L. Messi', 'height_cm']
```

```python
# get the weight of Cristiano Ronaldo 

df.loc['Cristiano Ronaldo', 'weight_kg']
```

```python
# get all rows inside the 'height_cm' column

df.loc[:, 'height_cm'] # the ':' gets all the elements in a row or column
```

```python
# get all columns that correspond to the index 'L. Messi'

df.loc['L. Messi', :]
```

### 2. Selecting with a list of values

```python
# get all data about L. Messi and Cristiano Ronaldo

df.loc[['L. Messi', 'Cristiano Ronaldo']]
```

```python
# get the height of L. Messi and Cristiano Ronaldo

df.loc[['L. Messi', 'Cristiano Ronaldo'], 'height_cm']
```

```python
# get the height and weight of L. Messi

df.loc['L. Messi', ['height_cm', 'weight_kg']]
```

```python
# get the height and weight of L. Messi and Cristiano Ronaldo

df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm', 'weight_kg']]
```

### 3. Selecting a range of data with a slice

start:stop:step (Note that contrary to usual python slices, both the start and the stop are included)

```python
# slice column labels

players =  ['L. Messi', 'Cristiano Ronaldo']
df.loc[players, 'age':'overall']
```

```python
# slice index labels

columns = ['age', 'dob', 'height_cm', 'weight_kg']

# get top1 and top10 player names


df.loc['L. Messi':'M. Salah', columns]
```

### 4. Selecting with conditions

```python
# one condition: select player with height above 180cm

columns = ['age', 'dob', 'height_cm', 'weight_kg']


df.loc[df['height_cm'] > 180, columns]
```

```python
# multiple conditions: select player with height above 180cm from Argentina


df.loc[(df['height_cm'] > 180) & (df['nationality'] == 'Argentina')]
```
