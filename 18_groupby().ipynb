```python
import pandas as pd

df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

```python
# using groupby method

group_by_lunch = df_exams.groupby('gender')
group_by_lunch.sum(numeric_only=True)

# the 'numeric_only=True' is very important when dealing with integers
```

```python
df_exams.groupby('gender').count()
```

```python
# agg function in pandas

df_exams.groupby('gender').agg({'math score': ['mean', 'max', 'min', 'count'], 'reading score': ['mean', 'max', 'min', 'count'], 'writing score':  ['mean', 'max', 'min', 'count']})
```

```python
# grouping multiple columns

df_exams.groupby(['gender', 'race/ethnicity']).mean(numeric_only=True)
```

```python
# shortcut with the describe() function

df_exams.groupby('gender').describe()
```
