### 3. adding a new column using 'assign()'

```python
score1 = np.random.randint(1, 100, size=1000)
score2 = np.random.randint(1, 100, size=1000)
```

```python
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
# using assign() to add the two columns to the dataframe

df_exams = df_exams.assign(score1 = series1, score2 = series2)
df_exams
```

### 4. adding a new column using insert()

```python
score3 = np.random.randint(1, 100, size=1000)
series3 = pd.Series(score3, index=np.arange(0,1000))
series3
```

```python
df_exams.insert(1, 'series3', series3)
```
