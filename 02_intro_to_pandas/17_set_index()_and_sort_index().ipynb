```python
import pandas as pd
import numpy as np
import random
```

```python
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

### set_index()

```python
# create non-repititive value for the index
new_index = np.arange(0, 1000)
```

```python
# shuffle the indexes using random

random.shuffle(new_index)
```

```python
# create a new column with the index

df_exams['new index'] = new_index
df_exams.head()
```

```python
# set the 'new index' column as index using 'set_index()'

df_exams.set_index('new index', inplace=True)
df_exams.head()
```

### sort_index()

```python
# sort the data frame by the new index using 'sort_index()'

df_exams.sort_index(inplace=True)
df_exams.head()
```
