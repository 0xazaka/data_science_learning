## 1. Creating a dataframe from an array

```python
import pandas as pd
import numpy as np
```

### Option 1: creating a dataframe with numpy


```python
data = np.array([[1,2], [2,3], [3,4], [4,5]])
print(data)
```

```python
df = pd.DataFrame(data, index = ['row1', 'row2', 'row3', 'row4'], columns=['col1', 'col2'])
df
```

### Option 2: creating a dataframe from a dictionary

```python
states = ['Lagos', 'Kaduna', ' Oyo', 'Cross River']
population = [25000000, 10000000, 9000000, 4000000]

dict_states = {'States': states, 'population': population}
```

```python
df_population = pd.DataFrame(dict_states)
df_population
```
