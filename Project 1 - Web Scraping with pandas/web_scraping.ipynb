## Read a .csv from Multiple URLs with pandas

Aim: to get the data from any of the leagues in england (premier, championship, league 1, league 2, conference) for the 2025/2026 season

```python
root = 'https://www.football-data.co.uk/mmz4281/'

leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

# loop through the leagues, read multiple csv and append it into a list

for league in leagues:
    df = pd.read_csv(root + '2526' + '/' + league + '.csv')
    frames.append(df)
```

### Now to extract data for multiple seasons


```python
for season in range(15,26):
    print(str(season) + str(season + 1))
```

```python
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []


# loop through the leagues, read multiple csv and append it into a list

for league in leagues:
    for season in range(20, 26):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + league + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
```

### 3. Organize all the data in a dictionary

```python
# create a dictionary with the original name of the leagues as the key

dict_countries = {
    'Spanish La liga' : 'SP1',
    'Spanish Segunda Division' : 'SP2',
    'German Bundesliga' : 'D1',
    'German 2. Bundesliga' : 'D2',
    'English Premier League' : 'E0',
    'English Championship' : 'E1'
}
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}

# creating list of leagues
leagues = ['E0', 'E1']



# loop through the leagues, read multiple csv and append it into a list

for league in dict_countries:
    frames = []
    for season in range(20, 26):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
